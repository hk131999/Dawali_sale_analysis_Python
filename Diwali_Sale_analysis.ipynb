```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Diwali Sales Data.csv")
data.info()
```

```python
data.describe()
```

```python
data.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data['Marital_Status']=data['Marital_Status'].replace([1,0],['Married','Not Married'])
```

```python
data
```

```python
fig,ax=plt.subplots(1,2)
fig.set_figwidth(40)
fig.set_figheight(10)
sns.set(font_scale=1)
sns.countplot(x=data.Gender,data=data,ax=ax[0])
sns.countplot(x='Gender',data=data,hue='Marital_Status',ax=ax[1])
Inf_Gender=pd.DataFrame()
Inf_Gender['Gender']=['F','M']
Inf_Gender['Count']=[data['Gender'].value_counts()[0],data['Gender'].value_counts()[1]]
print(data['Gender'].value_counts())
```

```python
Total_amount=pd.DataFrame()
Total_amount=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

print(Total_amount)
sns.barplot(x='Gender',y='Amount',data=Total_amount)


Inf_Gender['Amount']=Total_amount['Amount']
Inf_Gender['Avg Sale']=Inf_Gender['Amount']/Inf_Gender['Count']
Inf_Gender['Percentage_Contributation']=[(Inf_Gender['Amount'][0]/(Inf_Gender['Amount'].sum()))*100,(Inf_Gender['Amount'][1]/(Inf_Gender['Amount'].sum()))*100]
```

```python
ax=sns.countplot(x='Age Group', data=data,hue='Gender')
```

```python
Age_group_info=data.groupby(['Age Group','Gender'],as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
Age_group_info['Count']=(data.groupby(['Age Group','Gender'],as_index=False)['Gender'].size()).iloc[:,2]
Age_group_info['Avg Amount']=Age_group_info['Amount']/Age_group_info['Count']
Age_group_info['Percentage Contributation']=(Age_group_info['Amount']/Age_group_info['Amount'].sum())*100
```

```python
fig,ax=plt.subplots(1,2)
fig.set_figwidth(50)
fig.set_figheight(20)
sns.set(font_scale=1.5)
sns.barplot(x='Age Group',y='Amount',data=Age_group_info,hue='Gender',ax=ax[0])
sns.barplot(x='Gender',y='Amount',data=data,hue='Product_Category',ax=ax[1],palette='tab10')
Age_group_info.style.hide_index()
```

```python
fig,ax=plt.subplots(2,1)
fig.set_figwidth(40)
fig.set_figheight(30)
Region_sale_info=data.groupby(['Zone','State','Gender'], as_index=False)['Amount'].sum().sort_values(by='Zone',ascending=False)
Region_sale_info_pivot=Region_sale_info.pivot(index=['Zone','State'],columns='Gender',values='Amount')
sns.set(font_scale=1.5)
sns.barplot(x='Zone',y='Amount',data=data,ax=ax[0])
sns.barplot(x='State',y='Amount',data=data,ax=ax[1])
#ax[1].tick_params(axis='x', rotation=90)
Region_sale_info_pivot
```
